# 1. TRANSFORMER MODELS

The most basic object of transformers is "pipeline" witch enables to use a model for different tasks

## Pipeline object

```
from transformers import pipeline
```

### Sentiment analysis

```
classifier = pipeline("sentiment-analysis", device=device)
classifier(["The movie was very good", "I really disliked today´s match"])
```

### Zero shot classification

```
classifier = pipeline("zero-shot-classification", device=device)
classifier(["Today`s class at university was about international relations", "Tomorrow we have a meeting about the budget for the proyect"], candidate_labels=["education", "politics", "business"])
```

### Text generation

```
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M", device=device)
generator("The most beautiful town of the basque country is")
```

### Mask filler

```
filler = pipeline("fill-mask", device=device)
filler("The Algorithms and data estructures is a subject at <mask> that teach you a lot about automatization", top_k=2)
```
